```
dataset = keras.datasets.fashion_mnist
((training_images, training_identification),(test_images, test_identification)) = dataset.load_data()
```

```
#normalization
training_images = training_images/float(255)

#define the model with 2 layers
model = keras.Sequential([
  keras.layers.Flatten(input_shape=(28,28)),
  keras.layers.Dense(256, activation=tensorflow.nn.relu),
  keras.layers.Dense(128, activation=tensorflow.nn.relu),
  keras.layers.Dense(64, activation=tensorflow.nn.relu),
  keras.layers.Dense(10, activation=tensorflow.nn.softmax),  
])

#compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
  
#train the model
model.fit(training_images, training_identification)
```
